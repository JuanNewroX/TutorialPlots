### Importar los modulos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Cargando los datos para trabajar

```python

datos = pd.read_csv("playlistRandom.csv")
```

### Grafica - cantidad de canciones por playlist - Usando plot de pandas

```python
#Contamos la cantidad de canciones por playlist
datos.groupby('Playlist No.').count()['Track']
```

```python
#Realizamos el conteo y luego graficamos  le colocamos un tamaño para poder observar mejor la grafica figsize=(Ancho,Alto) pueden cambiar los valores para observar los cambios
datos.groupby('Playlist No.').count()['Track'].plot(kind='bar', figsize=(8,3))
```

### Grafica - cantidad de canciones por playlist - Usando Matplotlib

```python
#Con la función groupby agrupamos las playlist y contamos las canciones que aparecen en ella y seleccionamos los valores unicos
valores_x = datos.groupby(['Playlist No.']).Track.nunique().index
valores_y = datos.groupby(['Playlist No.']).Track.nunique().values

plt.bar(valores_x, valores_y)
```

### Grafica - cuantas veces se repiten las canciones en todas el archivo de playlist  - Usando plot de pandas

```python
#Realizamos el conteo de elementos en este caso las canciones
datos['Track'].value_counts()
```

```python
datos['Track'].value_counts().plot(kind='bar', figsize=(10,3))
```

### Grafica - cuantas veces se repiten las canciones en todas el archivo de playlist - Usando Matplotlib

```python
#Contamos la cantidad de veces que aparecen las canciones
#datos['Track'].value_counts()
#Los indices contienen los nombres de las canciones
valores_x = datos['Track'].value_counts().index
#Los values contienen la cantidad
valores_y = datos['Track'].value_counts().values

#Definimos un elemento de tipo figura para poder cambiar el tamanho de nuestra grafica figsize=(Ancho,Alto)
fig = plt.figure(figsize=(20,8)) 
#Colocamos los valores a graficar 
plt.bar(valores_x, valores_y, width = 0.8 )
#Rotamos las leyendas del eje x 90 grados
plt.xticks(rotation = 90)
plt.show()
```

### Grafica - cuantas veces aparecen los artistas en todas las playlist - Usando plot de pandas

```python
datos['Artist'].value_counts().plot(kind='bar')
```

### Grafica - cuantas veces aparecen los artistas en todas las playlist - Usando MatPlotLib

```python
#Contamos la cantidad de veces que aparecen los artistas
#datos['Artist'].value_counts()
#Los indices contienen los nombres de los artistas
valores_x = datos['Artist'].value_counts().index
#Los values contienen la cantidad
valores_y = datos['Artist'].value_counts().values

#Definimos un elemento de tipo figura para poder cambiar el tamanho de nuestra grafica figsize=(Ancho,Alto)
fig = plt.figure(figsize=(15,4)) 
#Colocamos los valores a graficar 
plt.bar(valores_x, valores_y, width = 0.8 )
#Rotamos las leyendas del eje x 90 grados
plt.xticks(rotation = 90)
plt.show()
```

### Grafica - Cuantas veces aparecen los artistas en cada playlist - Usando MatPlotlib

```python
## Utilizaremos un plot multiple de 5 columnas y 5 filas
fig, ax = plt.subplots(figsize=(20, 20), nrows=5, ncols=5)
## Cambiaremos el estilo de nuestra grafica  Pueden cambiar el codigo y probar dark_background (Investigar que otros estilos existen)
## plt.style.use('dark_background')
plt.style.use('ggplot')
plt.tight_layout()

posicion_pintar_grafica_x = 0
posicion_pintar_grafica_y = 0
plt.subplots_adjust(hspace = 1)
# Extraemos los datos de cada playlist para ello haremos un for de 1 a 25 
for i in range(1, 26):
    #Filtamos nuestros datos para extraer la playlist correspondiente
    datos_playlist_actual = datos[ datos['Playlist No.'] == i ]
    #Los indices contienen los nombres de los artistas filtrados
    valores_x = datos_playlist_actual['Artist'].value_counts().index
    #Los values contienen la cantidad
    valores_y = datos_playlist_actual['Artist'].value_counts().values

    ax [posicion_pintar_grafica_x, posicion_pintar_grafica_y].bar(valores_x, valores_y, color='r')
    ax [posicion_pintar_grafica_x, posicion_pintar_grafica_y].set_title("Grafica Canciones Playlist No. " + str(i))
    ax [posicion_pintar_grafica_x, posicion_pintar_grafica_y].tick_params(labelrotation=90)
    posicion_pintar_grafica_x = posicion_pintar_grafica_x + 1
    if posicion_pintar_grafica_x > 4:
        posicion_pintar_grafica_y = posicion_pintar_grafica_y + 1
        posicion_pintar_grafica_x = 0 


plt.show()
```

### Grafica de barras moviendo los ejes de agrupacion 

```python
## Creamos algunos datos de ejemplos 
estatura_hombres = [176, 164 , 169 , 121 , 135]
estatura_mujeres = [170, 160 , 150 ,110, 150]

#Creamos un vector para definir la posicion de cada barra 5 elementos, servira para poder seprara las barras entre hombre y mujer
xaxis = np.arange(5)
print(xaxis)

#Agregamos nuestros valores de hombres en el la posicion de nuestro vector xaxis
plt.bar(xaxis, estatura_hombres, width=0.45,  edgecolor="k", label="Male")
#Agregamos nuestros valores de mujeres en el la posicion en la posicion de nuestro xaxis + 0.45 (intenta cambiar el 0.45  observa que sucede) 
plt.bar(xaxis + 0.45, estatura_mujeres , width=0.45, edgecolor="k", color="pink",  label="Female" )
#Agramos valores a nuestro eje x para que se entienda mejor los datos
plt.xticks(xaxis + 0.45/2, {"Pareja 1", "Pareja 2", "Pareja 3", "Pareja 4", "Pareja 5"})

#Colocamos la leyenda de nuestro grafico en la localizacion 0, intenta cambiar el valor 0 por otros datos numericos
plt.legend(loc=0)
plt.show()
```

### Grafica de series

```python
datos_ejemplo_fechas = pd.read_csv("GOOG.csv") 
datos_ejemplo_fechas.head()
```

```python
#Creamos nuestros datos que se graficaran en el eje x en este caso la fecha pero inicialmente el formato es de tipo string lo debemos convertir a date usando to_datetime
indexTime=pd.to_datetime(datos_ejemplo_fechas['Date'])

#Definimos una figura con un tamanho de 8 y 5
fig_3 = plt.figure(figsize=(8,5))

#anhadimos nuestras valores de los ejes Y 
plt.plot(indexTime, datos_ejemplo_fechas['Low'])
plt.plot(indexTime, datos_ejemplo_fechas['High'])
plt.plot(indexTime, datos_ejemplo_fechas['Open'])
#Intenta agregar los otros valores de nuestro dataframe
plt.xticks(indexTime, rotation='vertical', size = 10 )

plt.show()
```
